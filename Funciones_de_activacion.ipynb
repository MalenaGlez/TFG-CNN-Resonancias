```python
import numpy as np
import matplotlib.pyplot as plt
```

## Función ReLu

```python
def ReLu(s):
  return np.maximum(0, s)
```

```python
x = np.linspace(-10, 10, 400)
y = ReLu(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, label='ReLu(x)')
plt.title('Función ReLU')
plt.xlabel('x')
plt.ylabel('ReLu(x)')
```

## Función tangente hiperbólica

```python
def tanh(s):
  return np.tanh(s)
```

```python
x = np.linspace(-10, 10, 400)
y = tanh(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, label='tanh(x)')
plt.title('Función tangente hiperbólica')
plt.xlabel('x')
plt.ylabel('tanh(x)')
```

## Función sigmoide logísitca

```python
def sigmoid(s):
  return 1 / (1 + np.exp(-s))
```

```python
x = np.linspace(-10, 10, 400)
y = sigmoid(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, label='sigmoid(x)')
plt.title('Función sigmoide logística')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
```
